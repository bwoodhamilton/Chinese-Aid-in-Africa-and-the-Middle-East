# K-Means Modeling 

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from sklearn.cluster import DBSCAN
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
aid = pd.read_csv('./aid_data/combined_data/aid_sums.csv')
```

```python
aid.head()
```

```python
ss = StandardScaler()
X_scaled = ss.fit_transform(aid[['world_bank_totals', 'chinese_aid_totals', 'usaid_aid', 'hdi_14', 'pr_score14', 'cl_score14', 'cpi_2014']])
```

```python
true_k = 8
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1, random_state=22)
model.fit(X_scaled)
```

```python
model.predict(X_scaled)
```

```python
aid['cluster'] = model.labels_
aid.head(25)
```

```python
silhouette_score(X_scaled, model.labels_)
```

```python
aid['cluster'].value_counts()
```

```python
aid.to_csv('./aid_data/combined_data/aid_clusters.csv', index=True)
```
